```python
import numpy as np
import torch
```

```python
inputs=np.array([[73,67,43],
                 [91,88,64],
                 [87,134,58],
                 [102,43,37],
                 [69,96,70]], dtype='float32')
print(inputs)
```

```python
targets=np.array([[56,70],
                  [81,101],
                  [119,133],
                  [22,37],
                  [103,119]], dtype='float32')
print(targets)
```

```python
inputs=torch.from_numpy(inputs)
targets=torch.from_numpy(targets)
print(inputs)
print(targets)
```

```python
w=torch.randn(2,3, requires_grad=True)
b=torch.randn(2 ,requires_grad=True)
print(w)
print(b)
```

```python
def model(x):
  return x @ w.t()+b
```

```python
preds=model(inputs)
print(preds)
```

```python
print(targets)
```

```python
diff=preds-targets
torch.sum(diff * diff)/diff.numel()
```

```python
def mse(t1 , t2):
  diff=t1-t2
  return torch.sum(diff * diff)/diff.numel()
```

```python
loss=mse(preds,targets)
print(loss)
```

```python
loss.backward()
```

```python
print(w)
print(w.grad)
```

```python
print(b)
print(b.grad)
```

```python
print(w)
w.grad
```

```python
with torch.no_grad():
  w-=w.grad*1e-5
  b-=b.grad*1e-5
```

```python
w, b
```

```python
preds=model(inputs)
loss=mse(preds,targets)
print(loss)
```

```python
w.grad.zero_()
b.grad.zero_()
print(w.grad)
print(b.grad)
```

```python
for i in range(500):
  preds=model(inputs)
  loss=mse(preds,targets)
  loss.backward()
  with torch.no_grad():
    w-=w.grad*1e-5
    b-=b.grad*1e-5
    w.grad.zero_()
    b.grad.zero_()
```

```python
preds=model(inputs)
loss=mse(preds,targets)
print(loss)
```
